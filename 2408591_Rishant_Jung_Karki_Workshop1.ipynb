Display Fibonacci series up to 10 terms.

```
a, b = 0, 1
count = 0
terms = 10

print("Fibonacci series:")
while count < terms:
    print(a, end=" ")
    nth = a + b
    a = b
    b = nth
    count += 1
```

Find the factorial of a given number. Ask the user to provide the number.

```
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

num_str = input("Enter a non-negative integer to find its factorial: ")
try:
    num = int(num_str)
    result = factorial(num)
    print(f"The factorial of {num} is {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

Write a program to print a right angle triangle using the for loop.

```
rows_str = input("Enter the number of rows for the right-angle triangle: ")

try:
    rows = int(rows_str)
    if rows <= 0:
        print("Please enter a positive integer for the number of rows.")
    else:
        for i in range(1, rows + 1):
            for j in range(i):
                print("*", end="")
            print()
except ValueError:
    print("Invalid input. Please enter an integer.")
```

Exercise on Functions

Make a simple calculator that can add, subtract, multiply and divide two numbers using function.

```
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    return x / y

print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

while True:
    choice = input("Enter choice(1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter numbers only!")
            continue

        if choice == '1':
            print(num1, "+", num2, "=", add(num1, num2))
        elif choice == '2':
            print(num1, "-", num2, "=", subtract(num1, num2))
        elif choice == '3':
            print(num1, "*", num2, "=", multiply(num1, num2))
        elif choice == '4':
            print(num1, "/", num2, "=", divide(num1, num2))

        # Ask if the user wants another calculation
        next_calculation = input("Let's do next calculation? (yes/no): ")
        if next_calculation == "no":
            break
    else:
        print("Invalid Input")
```

Exercise on List and Dictionaries

Remove the first duplicate element of a list

```
def remove_first_duplicate(input_list):
    seen = set()
    duplicate_found = False
    for item in input_list:
        if item in seen:
            # Found the first duplicate
            input_list.remove(item) # Remove the first occurrence of this item
            duplicate_found = True
            break
        else:
            seen.add(item)

    if not duplicate_found:
        print("No duplicate elements found in the list.")
    return input_list

# Example usage:
my_list = [1, 2, 3, 2, 4, 5, 3, 6]
print("Original list:", my_list)
modified_list = remove_first_duplicate(my_list.copy()) # Use .copy() to avoid modifying the original list directly in the function
print("List after removing the first duplicate:", modified_list)

my_list_no_duplicates = [10, 20, 30, 40]
print("\nOriginal list:", my_list_no_duplicates)
modified_list_no_duplicates = remove_first_duplicate(my_list_no_duplicates.copy())
print("List after removing the first duplicate:", modified_list_no_duplicates)
```

Write a Python program to create a new dictionary by extracting the mentioned keys from the below dictionary.
sample_dict = { "name": "Anderson", "age": 35, "salary": 5000, "city": "London"}
keys = ["name", "salary"]

```
sample_dict = { "name": "Anderson", "age": 35, "salary": 5000, "city": "London"}
keys = ["name", "salary"]

new_dict = {k: sample_dict[k] for k in keys if k in sample_dict}

print("Original dictionary:", sample_dict)
print("Keys to extract:", keys)
print("New dictionary:", new_dict)
```

```
import pandas as pd
import numpy as np
import plotly.express as px

# STEP 1: CREATE DATA
# (We are creating 121 rows that match our real-world stats)
n = 121
np.random.seed(42)

temp = np.random.uniform(15, 43, n)
price = np.random.lognormal(mean=3.7, sigma=0.7, size=n) + 5
price = np.clip(price, 9, 250)
df = pd.DataFrame({'Temp': temp, 'Price_per_kg': price})
```

```
# 1. SCATTER PLOT: Does Temp affect Price?
fig1 = px.scatter(df, x="Temp", y="Price_per_kg", trendline="ols",
                  title="1. Relationship: Temperature vs Price")
fig1.show()
```

```
# 2. HISTOGRAM: How are prices spread out?
fig2 = px.histogram(df, x="Price_per_kg", title="2. Price Distribution (Spread)")
fig2.show()
```

```
# 3. BOX PLOT: Where are the weird 'outlier' prices?
fig3 = px.box(df, y="Price_per_kg", points="all", title="3. Finding Outliers")
fig3.show()
```

```
# 4. BAR CHART: Average price in Simple Categories
df['Temp_Group'] = pd.cut(df['Temp'], bins=3, labels=['Cold', 'Medium', 'Hot'])
avg_price = df.groupby('Temp_Group', observed=False)['Price_per_kg'].mean().reset_index()

fig4 = px.bar(avg_price, x='Temp_Group', y='Price_per_kg', color='Temp_Group',
              title="4. Average Price by Category")
fig4.show()
```

```
# 5. HEATMAP: Where is the most data concentrated?
fig5 = px.density_heatmap(df, x="Temp", y="Price_per_kg", title="5. Data Density Map")
fig5.show()
```
